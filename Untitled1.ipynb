```python
class companyshare(object):
    def __init__(self,text, title, description):
        self.text = text
        self.title = title
        self.description = description
    
    def toline(self):
        n_str = 'text: %s\narticle title: %s\narticle description: %s\n' % (self.text, self.title, self.description)
        return n_str
```

```python
res_lst = []
header = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_4) AppleWebKit/605.1.15 (KHTML, like Gecko) Version/11.1 Safari/605.1.15'}

def text_process(ntxturl):
    text = ntxturl.text
    url = ntxturl.url
    
    if (url == None):
        share = companyshare(text,'No url in this post',' ')
        res_lst.append(share)
    else:
        try:
            res = requests.get(url)
            content = res.text
        
            title_begin = content.find('<title')
            title_begin = content.find('>',title_begin)
            title_end = content[title_begin:].find('</')
            title = content[title_begin+1:title_begin+title_end].strip(' ')

            des_begin = content.find('name="description"')
            l = len('name="description" content="')
            if (des_begin == -1):
                des_begin = content.find('og:description')
                l = len('og:description" content="')
            des_end = content[des_begin:].find('>')
            des = content[des_begin+l:des_begin+des_end].strip(' ')
            if (des_begin == -1):
                des = 'Page description not found!'
        
            share = companyshare(text, title, des)
            res_lst.append(share)
        
        except requests.exceptions.ConnectionError:
#         print('Cannot reach this page!\n')
            share = companyshare(text,'Page cannot be reached',' ')
            res_lst.append(share)
```

```python
acc = 1
for text in text_lst:
    text_process(text)
    print(acc)
    acc = acc+1
print('Done!')
```
